```python
import sys
sys.path.append("music-transcription/")
import datasets
import visualization
import matplotlib.pyplot as plt
import numpy as np
```

```python
def preload_fn(aa):
    aa.annotation = datasets.Annotation.from_time_series(*aa.annotation, hop_samples=512)
    aa.audio.load_resampled_audio(44100)

```

```python
adc2004_test = datasets.mirex05.prepare(preload_fn)
annot = adc2004_test[1].annotation
# datasets.common.
ref = annot.notes_mf0[1000:1500]
est = np.squeeze(annot.notes[1000:1500])
est[20:35] = 50
est[126:135] = 46
est[175:185] = 60+np.sin(np.arange(10)*0.2)
est = datasets.common.melody_to_multif0(est)
fig, axs = plt.subplots(1, 1, sharex=True, sharey=False, squeeze=False, figsize=(6, 4))
axs = axs[:,0]
visualization.draw_notes_on_ax(ref, est, ax=axs[0], ms=5)
plt.ylim(40,70)
fig.savefig("chyba_FA.svg", bbox_inches="tight")
```

```python
adc2004_test = datasets.mirex05.prepare(preload_fn)
annot = adc2004_test[1].annotation
# datasets.common.
ref = annot.notes_mf0[1000:1500]
est = np.squeeze(annot.notes[1000:1500])
est[240:280:3] -= 8
est[241:280:4] += 4
est[200:240] += 7
est[50:80] -= 1
est[135:145] = 45
est[145:175] = 47

est = datasets.common.melody_to_multif0(est)
fig, axs = plt.subplots(1, 1, sharex=True, sharey=False, squeeze=False, figsize=(6, 4))
axs = axs[:,0]
visualization.draw_notes_on_ax(ref, est, ax=axs[0], ms=5)
plt.ylim(40,70)
fig.savefig("chyba_RPA.svg", bbox_inches="tight")
```
